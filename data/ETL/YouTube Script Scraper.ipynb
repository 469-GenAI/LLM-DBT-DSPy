## Use Youtube Data's Search API to query "Sharktank" + "Company Name"

```python
import requests
import json
import time


def getYoutubeVideoIds(listOfKeywords):
  output = {}
  YT_SEARCH_API_BASE_URL = "https://www.googleapis.com/youtube/v3/search?part=snippet&maxResults=1&type=video&q="
  for keyword in search_keywords:
    url = YT_SEARCH_API_BASE_URL + keyword + "SharkTank" + "&key=" + YT_DATA_API_KEY
    response = requests.get(url)
    data = json.loads(response.text)
    if data and data['items']:
      video_id = data['items'][0]['id']['videoId']
      output[keyword] = video_id
      # print(f"Keyword: {keyword}, Video ID: {video_id}")
    else:
      output[keyword] = None
      # print(f"No video found for keyword: {keyword}")
    time.sleep(3) # Add a 3-second pause between requests
  return output

video_id_dict = getYoutubeVideoIds(search_keywords)
print(video_id_dict)
```

```python
# Youtube to TranscriptAPI
import os
import requests
import json
import time

TRANSCRIPTAPI_API_KEY = userdata.get("TRANSCRIPTAPI_API_KEY")
def transcribeYoutubeVideo(videoId):
  """
  Takes in a youtube video id and returns the transcript list of the video.
  """
  url = 'https://transcriptapi.com/api/v2/youtube/transcript'
  params = {'video_url': videoId, 'format': 'json'}
  response = requests.get(url, params=params, headers={'Authorization': 'Bearer ' + TRANSCRIPTAPI_API_KEY}, timeout=10)
  response.raise_for_status()
  transcript_array = response.json()['transcript']
  # print(transcript_array)
  return transcript_array


def concatTranscriptText(transcript_array):
  transcript = ""
  for chunk in transcript_array:
    transcript += chunk['text'] + " "
  return transcript

def getTranscript(video_id_dict):
  transcript_dict = {}
  ctr = 0;
  for keyword, video_id in video_id_dict.items():
    try:
      transcript_array = transcribeYoutubeVideo(video_id)
      transcript = concatTranscriptText(transcript_array)
      transcript_dict[keyword] = transcript
      ctr += 1
      if ctr % 10 == 0:
        print(f"Processed {ctr} videos.")
    except requests.exceptions.RequestException as e:
      print(f"Error transcribing video for keyword '{keyword}' (Video ID: {video_id}): {e}")
      transcript_dict[keyword] = None # Or handle the error in another way
  return transcript_dict

# transcript_dict = getTranscript(video_id_dict)
transcript_dict = getTranscript(yt_videos_data)
# transcript_dict = getTranscript(videos)
# print(transcript_dict)
print(len(transcript_dict.keys()))
```
