Extract company names and video IDs from the descriptions in the "video_ids_&_description.json" file using the `gpt-4o-mini-2024-07-18` model, processing in batches of 10 with throttling, filtering for descriptions that sound like a startup pitch and removing compilations. The final output should be a dictionary mapping company names to video IDs.

## Define the processing function

### Subtask:
Create a function that takes a batch of video ID and description pairs, constructs prompts for the GPT model to identify startup pitches and extract company names, and returns the filtered results.


```python
import openai
import time
import json # Import the json library

# Initialize the OpenAI client with the API key
openai.api_key = OPENAI_API_KEY

def process_batch(batch_data):
    """
    Processes a batch of video ID and description pairs using GPT to identify startup pitches
    and extract company names.

    Args:
        batch_data: A dictionary where keys are video IDs and values are descriptions.

    Returns:
        A dictionary containing video IDs and extracted company names for startup pitches.
    """
    refined_results = {}
    for video_id, description in batch_data.items():
        # Skip entries with None description
        if description is None:
            continue

        # Exclude known compilation videos based on keywords in description
        if any(keyword in description.lower() for keyword in ["top 3", "biggest deals", "moments", "shark tank"]):
            continue

        prompt = f"""
        Analyze the following video description to determine if it describes a startup pitch on a show like Shark Tank.
        If it is a startup pitch, identify and extract the company name being pitched.
        Exclude descriptions that are compilations or highlight reels (e.g., "Top 10 moments", "Biggest deals", "Best of Shark Tank").
        Output the result in a JSON format with two keys:
        "is_pitch": boolean (true if it's a startup pitch, false otherwise)
        "videoId": string (the corresponding videoId)
        "company_name": string (the extracted company name, or null if not a pitch or no company name found)

        Video ID: {video_id}
        Description: {description}
        """

        try:
            response = openai.chat.completions.create(
                model="gpt-4o-mini-2024-07-18",
                response_format={"type": "json_object"},
                messages=[
                    {"role": "system", "content": "You are a helpful assistant that analyzes video descriptions."},
                    {"role": "user", "content": prompt}
                ]
            )
            result = json.loads(response.choices[0].message.content)

            if result.get("is_pitch") and result.get("company_name"):
                refined_results[video_id] = result["company_name"]
        except Exception as e:
            print(f"Error processing video ID {video_id}: {e}")
            # Implement a small delay before retrying or moving to the next item
            time.sleep(1) # Simple throttling

    return refined_results

print("Function process_batch defined.")
```

```python
import time

company_video_ids = {}
batch_size = 10
video_items = list(video_data.items())

for i in range(0, len(video_items), batch_size):
    batch_items = dict(video_items[i:i + batch_size])
    print(f"Processing batch {int(i/batch_size) + 1}...")
    batch_results = process_batch(batch_items)
    company_video_ids.update(batch_results)
    print(f"Finished batch {int(i/batch_size) + 1}. Waiting before next batch...")
    time.sleep(5) # Throttling

print(f"Finished processing all batches. Total companies found: {len(company_video_ids)}")
```

# Task
Analyze the length of the "Full_Pitch" field in the "existing_refinend_pitches.json" file, calculate the average length, and identify pitches that are significantly shorter than the average.

## Load the json data

### Subtask:
Load the data from the "existing_refinend_pitches.json" file into a Python dictionary or list of dictionaries.


```python
import json

file_path = "existing_refined_pitches.json"
with open(file_path, 'r') as f:
    pitches_data = json.load(f)

print(f"Loaded {len(pitches_data)} pitch entries.")
```

## Extract and measure pitch lengths

### Subtask:
Iterate through the loaded data and extract the "Full_Pitch" value for each entry. Measure the length of each extracted pitch string.


```python
pitch_lengths = []

for pitch_entry in pitches_data:
    full_pitch = pitch_entry.get("Full_Pitch")
    if isinstance(full_pitch, str):
        pitch_lengths.append(len(full_pitch))

print(f"Extracted lengths for {len(pitch_lengths)} pitches.")
```

## Analyze pitch lengths

### Subtask:
Calculate descriptive statistics for the pitch lengths (e.g., median, quartiles) to get a better understanding of the distribution and identify a suitable threshold for "significantly shorter" pitches, potentially considering the user's suggestion of around 1.2k characters.


```python
import numpy as np

median_pitch_length = np.median(pitch_lengths)
q1_pitch_length = np.percentile(pitch_lengths, 25)
q3_pitch_length = np.percentile(pitch_lengths, 75)

print(f"Median pitch length: {median_pitch_length}")
print(f"First quartile (25th percentile): {q1_pitch_length}")
print(f"Third quartile (75th percentile): {q3_pitch_length}")
```

## Identify pitches below the threshold

### Subtask:
Based on the calculated descriptive statistics, identify pitches that are significantly shorter than the median length (1345 characters) and might need further attention.


```python
significantly_shorter_pitches = []
exceptionally_longer_pitches = []

for pitch_entry in pitches_data:
    full_pitch = pitch_entry.get("Full_Pitch")
    if isinstance(full_pitch, str):
        if len(full_pitch) < q1_pitch_length:
            significantly_shorter_pitches.append({"Product": pitch_entry.get("Product"), "Length": len(full_pitch)})
        elif len(full_pitch) > q3_pitch_length:
             exceptionally_longer_pitches.append({"Product": pitch_entry.get("Product"), "Length": len(full_pitch)})

print(f"Found {len(significantly_shorter_pitches)} pitches significantly shorter than the first quartile ({q1_pitch_length:.2f} characters):")
for pitch in significantly_shorter_pitches:
    print(f"- {pitch['Product']}: {pitch['Length']} characters")

print(f"\nFound {len(exceptionally_longer_pitches)} pitches exceptionally longer than the third quartile ({q3_pitch_length:.2f} characters):")
for pitch in exceptionally_longer_pitches:
    print(f"- {pitch['Product']}: {pitch['Length']} characters")
```
